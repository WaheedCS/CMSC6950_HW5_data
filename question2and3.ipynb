```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the iterations data file
iterations_file = np.load("iterations.npz")
iteration_counts = iterations_file['iteration_counts']

#### Plot histogram with kernel density estimate
plt.figure(figsize=(10, 6))
sns.histplot(iteration_counts, kde=True, bins=20, color='skyblue')
plt.title("Histogram of Iteration Counts with Kernel Density Estimate")
plt.xlabel("Iteration Counts")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

## heights data
heights_file = np.load("heights.npz")
X = heights_file['X']
Y = heights_file['Y']
height_25 = heights_file['height_25']
height_50 = heights_file['height_50']
height_75 = heights_file['height_75']
height_100 = heights_file['height_100']

# Plot heat maps for each time step
time_steps = [0.25, 0.5, 0.75, 1.0]
heights = [height_25, height_50, height_75, height_100]
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

for i, ax in enumerate(axes.flatten()):
    heat = heights[i]
    c = ax.pcolormesh(X, Y, heat, shading='auto', cmap='viridis')
    fig.colorbar(c, ax=ax, orientation='vertical')
    ax.set_title(f"Water Height at t = {time_steps[i]}")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

## Loading RR data file
RR_data_file = np.load("RR_data.npz")

## Reynolds numbers and grid resolutions to plot
reynolds_numbers = [5000, 10000, 20000]
grid_resolutions = ['nref4', 'nref5', 'nref6', 'nref7']

## Ploting line graphs for each Reynolds number
plt.figure(figsize=(14, 10))
for Re in reynolds_numbers:
    plt.subplot(3, 1, reynolds_numbers.index(Re) + 1)
    
    ## Ploting each grid resolution for the current Reynolds number
    for res in grid_resolutions:
        rr_key = f"RR{Re}_{res}"
        time_key = f"times_{res}"
        
        rr_data = RR_data_file[rr_key]
        times = RR_data_file[time_key]
        
        plt.plot(times, rr_data, label=f"Resolution {res}")
    
    plt.title(f"Reconnection Rate at Reynolds Number {Re}")
    plt.xlabel("Time")
    plt.ylabel("Reconnection Rate")
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```
